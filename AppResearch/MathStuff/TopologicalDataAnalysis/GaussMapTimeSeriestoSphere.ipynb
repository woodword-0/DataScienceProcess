```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate a cyclical time series
time = np.linspace(0, 10, 1000)
amplitude = np.sin(time)
amplitude1 = np.tan(time)

# Create a rolling window
window_size = 2
X = np.column_stack([np.roll(amplitude, i) for i in range(window_size)])
Y = np.column_stack([np.roll(amplitude1, i) for i in range(window_size)])
# Normalize to unit length (map onto a circle)
X = X / np.linalg.norm(X, axis=1)[:, np.newaxis]
Y = Y / np.linalg.norm(Y, axis=1)[:, np.newaxis]

# Convert to spherical coordinates
rho = np.linalg.norm(X, axis=1)
phi = np.arctan2(X[:, 1], X[:, 0])
rho1 = np.linalg.norm(Y, axis=1)
phi1 = np.arctan2(Y[:, 1], Y[:, 0])

# Plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='polar')
ax.scatter(phi, rho)
ax.scatter(phi1, rho1, c = 'r')
plt.show()
```

# Plotting a time series on the sphere
